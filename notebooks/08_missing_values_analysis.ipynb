```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('../data/processed/train_preprocessed.csv')
df.head()
```

```python
# check relationship between Cabin and FamilySize
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Cabin', hue='FamilySize', palette='viridis')
plt.title('Cabin vs Family Size')
plt.xlabel('Cabin')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Family Size')
plt.tight_layout()
plt.show()
```
